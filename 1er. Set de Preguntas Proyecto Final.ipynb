####  ¿Cual es el estado con más laboratorios y cual es la probabilidad que tenga mas de el 200 encuestados?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
homehealth=pd.read_csv("homehealth.csv")
bueno=50;
info=homehealth[homehealth['Percentr']==bueno]
totales=sum(homehealth['Sratings']==1)
laboratorios=sum(info['Sratingh']==1)
probabilidad=((laboratorios/1.0)/(totales/1.0))*100
info2=homehealth[homehealth['Percentr']==bueno]
lab2=sum(info2['Sratinghs']==1)
probabilidad2=((lab2/1.0)/(totales/1.0))*100
probabilidadtotal=probabilidad+probabilidad2;
print("probabilidad de que un laboratorio tenga rating menor de "+ str(bueno) +" puntos es de "+ str(probabilidadtotal)[0:4] + "%")
pd.crosstab(index=homehealth['State'],
            columns=homehealth["Percentr"], margins=True)
```
